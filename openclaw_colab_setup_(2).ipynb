```python
# Create .env file with Groq API configuration
import os

env_content = f"""GROQ_API_KEY={os.environ['GROQ_API_KEY']}
ANTHROPIC_API_KEY=
OPENAI_API_KEY=
"""

with open('.env', 'w') as f:
    f.write(env_content)

print("✓ .env file created with Groq API key")
```

```python
# Create OpenClaw configuration file
import json

config = {
    "gateway": {
        "mode": "local",
        "port": 3000
    },
    "model": {
        "provider": "groq",
        "name": "llama-3.3-70b-versatile"
    }
}

# Create the config directory if it doesn't exist
!mkdir -p ~/.config/openclaw

# Write the config file
with open(os.path.expanduser('~/.config/openclaw/config.json'), 'w') as f:
    json.dump(config, f, indent=2)

print("✓ OpenClaw configuration created")
print("\nConfiguration:")
print("- Mode: local")
print("- Provider: Groq")
print("- Model: llama-3.3-70b-versatile")
print("- Port: 3000")
```

```python
# Start ngrok tunnel in the background
import subprocess
import time
import requests
import json

# Start ngrok
ngrok_process = subprocess.Popen(
    ['ngrok', 'http', '3000', '--log=stdout'],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE
)

# Wait for ngrok to start
time.sleep(3)

# Get ngrok public URL
try:
    response = requests.get('http://localhost:4040/api/tunnels')
    tunnels = response.json()['tunnels']
    public_url = tunnels[0]['public_url']
    print(f"\n{'='*60}")
    print(f"🌐 ngrok tunnel started successfully!")
    print(f"   Public URL: {public_url}")
    print(f"   Dashboard: http://localhost:4040")
    print(f"{'='*60}\n")
    print("⏳ Waiting for OpenClaw to start on port 3000...")
except Exception as e:
    print(f"⚠️  ngrok started but URL not yet available")
    print(f"   Check http://localhost:4040 for the ngrok dashboard")
```

```python
# Check if OpenClaw is running and display access information
import requests
import time

def check_openclaw():
    """Check if OpenClaw is responding on port 3000"""
    try:
        response = requests.get('http://localhost:3000', timeout=2)
        return response.status_code in [200, 404]  # Both indicate server is running
    except:
        return False

def get_ngrok_url():
    """Get the ngrok public URL"""
    try:
        response = requests.get('http://localhost:4040/api/tunnels')
        tunnels = response.json()['tunnels']
        return tunnels[0]['public_url']
    except:
        return None

print("🔍 Checking OpenClaw status...\n")

# Check local server
if check_openclaw():
    print("✅ OpenClaw is running on http://localhost:3000")
else:
    print("❌ OpenClaw is NOT running on port 3000")
    print("   Run the server start cell below to launch it")

# Check ngrok
public_url = get_ngrok_url()
if public_url:
    print(f"✅ ngrok tunnel is active")
    print(f"\n{'='*60}")
    print(f"🌍 ACCESS OPENCLAW EXTERNALLY:")
    print(f"   {public_url}")
    print(f"{'='*60}")
    print(f"\n📊 ngrok Dashboard: http://localhost:4040")
else:
    print("❌ ngrok tunnel is NOT active")
    print("   Run the ngrok cell above to start it")
```

```python
# Start OpenClaw server
print("🦞 Starting OpenClaw gateway...\n")
print("WHAT TO LOOK FOR:")
print("  ✅ 'Gateway listening on port 3000' or similar message")
print("  ✅ No error messages about missing config")
print("  ✅ The cell keeps running (doesn't exit immediately)\n")
print("The server will run until you stop this cell.")
print("Once started, run the status check cell above to get your ngrok URL.\n")
print("="*60)

# Start the gateway on port 3000
!pnpm openclaw gateway --port 3000 --verbose
```

## Troubleshooting

If you need to restart the server:
1. Stop the cell above (click the stop button)
2. Run it again

To check ngrok status:
- Visit http://localhost:4040 in a new Colab output or external browser

To use a different Groq model:
- Modify the `MODEL_NAME` in the .env creation cell
- Available models: `llama-3.3-70b-versatile`, `llama-3.1-8b-instant`, `mixtral-8x7b-32768`, etc.

To use different AI providers:
- OpenClaw supports Anthropic, OpenAI, Groq, and others
- Set the appropriate API key and MODEL_PROVIDER in the .env cell
